```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# station metadata
stn_ids = pd.read_fwf('http://noaa-ghcn-pds.s3.amazonaws.com/ghcnd-stations.txt', header=None, infer_nrows=1000)
stn_ids.columns = ['ID','LAT','LON','ELEV','UKN','NAME','GSN','WBAN']

# Shreveport, LA stations
shreveport_stations = stn_ids[stn_ids['NAME'].str.contains('SHREVEPORT', case=False, regex=False)]
shreveport_stations
```

```python
# Data for Shreveport
df = pd.read_parquet(
    "s3://noaa-ghcn-pds/parquet/by_station/STATION=USW00013957/",
    storage_options={"anon": True}
)

df['DATE'] = pd.to_datetime(df['DATE'].astype(str), format='%Y%m%d')
df = df.set_index('DATE').sort_index()

# Elements
df_tmax = df[df['ELEMENT'] == 'TMAX']
df_tmin = df[df['ELEMENT'] == 'TMIN']
```

```python
def daily_temperature_stats(df_tmax, df_tmin):
    # 1991–2020 Temp Data
    df_tmax = df_tmax[(df_tmax.index.year >= 1991) & (df_tmax.index.year <= 2020)]
    df_tmin = df_tmin[(df_tmin.index.year >= 1991) & (df_tmin.index.year <= 2020)]

    df_tmax['DAY'] = df_tmax.index.strftime('%m-%d')
    df_tmin['DAY'] = df_tmin.index.strftime('%m-%d')

    stats = pd.DataFrame(index=sorted(df_tmax['DAY'].unique()))
    stats['record_max_temp'] = df_tmax.groupby('DAY')['DATA_VALUE'].max() / 10
    stats['average_max_temp'] = df_tmax.groupby('DAY')['DATA_VALUE'].mean() / 10
    stats['record_min_temp'] = df_tmin.groupby('DAY')['DATA_VALUE'].min() / 10
    stats['average_min_temp'] = df_tmin.groupby('DAY')['DATA_VALUE'].mean() / 10

    return stats

stats = daily_temperature_stats(df_tmax, df_tmin)
print(stats)
#stats.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

def plot_shreveport(stats, df_tmax, df_tmin, year):
    df_year_max = df_tmax[df_tmax.index.year == year].copy()
    df_year_min = df_tmin[df_tmin.index.year == year].copy()

    df_year_max['DAY'] = df_year_max.index.strftime('%m-%d')
    df_year_min['DAY'] = df_year_min.index.strftime('%m-%d')

    actual_max = df_year_max.groupby('DAY')['DATA_VALUE'].mean() / 10
    actual_min = df_year_min.groupby('DAY')['DATA_VALUE'].mean() / 10

    # Converting calendar days to datetime 
    days = pd.to_datetime('2020-' + stats.index, format='%Y-%m-%d')
    actual_max_days = pd.to_datetime('2020-' + actual_max.index, format='%Y-%m-%d')
    actual_min_days = pd.to_datetime('2020-' + actual_min.index, format='%Y-%m-%d')

    fig, ax = plt.subplots(figsize=(15, 6))

    ax.plot(days, stats['record_max_temp'], label='Record High', color='red')
    ax.plot(days, stats['average_max_temp'], label='Avg High', color='orange')
    ax.plot(days, stats['record_min_temp'], label='Record Low', color='blue')
    ax.plot(days, stats['average_min_temp'], label='Avg Low', color='skyblue')
    ax.plot(actual_max_days, actual_max.values, '--', label=f'{year} High', color='darkred')
    ax.plot(actual_min_days, actual_min.values, '--', label=f'{year} Low', color='darkblue')

    # Making x-axis for readable
    ax.xaxis.set_major_locator(mdates.MonthLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))
    ax.set_xlim([pd.to_datetime('2020-01-01'), pd.to_datetime('2020-12-31')])

    ax.set_title(f"Shreveport, LA Temperatures in {year}")
    ax.set_xlabel("Month")
    ax.set_ylabel("Temperature (°C)")
    ax.legend()
    plt.tight_layout()
    plt.show()
    
# 2020 Data
plot_shreveport(stats, df_tmax, df_tmin, 2020)
```
